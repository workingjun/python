```python
root = Tk()
root.title("나도 GUI") ##제목 설정
root.geometry("640x480") ##가로 * 세로

txt=Text(root, width=30, height=5)
txt.pack()

txt.insert(END, "글자를 입력하세요")

e=Entry(root, width=30) ##엔터 안됨 -> 여러줄 안됨 
e.pack()
e.insert(0, "한줄만 입력하세요") ##처음위치 

def btncmd():
    print(txt.get("1.0", END)) #"1.0"은 라인 1부터 가져오기 
                               #0은 컬럼0부터 가져오기
    print(e.get())

    #내용 삭제 
    txt.delete("1.0", END)
    e.delete(0, END)
    
    
btn=Button(root, text="클릭", command=btncmd)
btn.pack()

root.mainloop()
```

```python
from tkinter import * 


root = Tk()
root.title("나도 GUI") ##제목 설정
root.geometry("640x480") ##가로 * 세로

def create_new_file():
    print("새 파일을 만듭니다")

menu= Menu(root)

#파일 메뉴
menu_file=Menu(menu, tearoff=0)
menu_file.add_command(label="New file", command=create_new_file)
menu_file.add_command(label="New window")
menu_file.add_separator()
menu_file.add_command(label="open file...")
menu_file.add_separator()

menu_file.add_command(label="save all", state="disable")
menu_file.add_separator()
menu_file.add_command(label="Exit", command=root.quit)
menu.add_cascade(label="File", menu=menu_file)

#에딧 메뉴 (빈값)
menu.add_cascade(label="Edit")

#랭기지 메뉴 추가
menu_lag=Menu(menu, tearoff=0)
menu_lag.add_radiobutton(label="python")
menu_lag.add_radiobutton(label="Java")
menu_lag.add_radiobutton(label="C++")
menu.add_cascade(label="Language", menu=menu_lag)


menu_view=Menu(menu, tearoff=0)
menu_view.add_checkbutton(label= "show Minimap")
menu.add_cascade(label="View", menu=menu_view)

root.config(menu=menu)
root.mainloop()
```

```python
from tkinter import * 
import tkinter.messagebox as mbox

root = Tk()
root.title("나도 GUI") ##제목 설정
root.geometry("640x480") ##가로 * 세로

def info():
    mbox.showinfo("알림", "정상적으로 예매 완료")

def warn():
    mbox.showwarning("경고", "해당 좌석은 매진")

def error():
    mbox.showerror("에러", "결제 오류 발생")

def okcancel():
    mbox.askokcancel("확인/취소", "해당 좌석은 유아동반석입니다. 예매하시겠습니까?")

def retrycancel():
    response = mbox.askretrycancel("재시도 /취소", "일시적인 오류입니다. 다시 시도")
    if response==1: ##재시도
        print("재시도")
    elif response==0: ##취소 
        print("취소")
   

def yesno():
    mbox.askyesno("예/아니오", "해당 좌석은 역방향입니다. 예매 하시겠습니까?")

    
def yesnocancel():
    response = mbox.askyesnocancel(title=None, message="예매내역이 저장되지 않았습니다 \n 저장후 프로그램을 종료하시겠습니까?")

    if response==1: ##재시도
        print("예")
    elif response==0: ##취소 
        print("아니오")
    else:
        print("취소")


Button(root, command=info, text="알림").pack()
Button(root, command=warn, text="경고").pack()
Button(root, command=warn, text="에러").pack()
Button(root, command=okcancel, text="확인 취소").pack()
Button(root, command=retrycancel, text="재시도 취소").pack()
Button(root, command=yesno, text="예 아니오").pack()

Button(root, command=yesnocancel, text="예 아니오 취소").pack()



root.mainloop()
```
