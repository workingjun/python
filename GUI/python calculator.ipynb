계산기 프로그램

```python
from tkinter import *
from numpy import *

def factorial(n):
    if n==1:
        return 1
    else:
        return n * factorial(n-1)
    
def button_click(text):
    global display

    if text == 'C':
        display.delete(1.0, END)
    elif text == '=':
        try:
            text_calc = display.get("1.0", END)
            
            if text_calc.endswith('!\n'):
                text_calc = text_calc.replace('!\n', '')
                fac_num = eval(text_calc)

                result = factorial(fac_num)
                
            else:
                if 'ln' in text_calc:
                    text_calc=text_calc.replace('ln', 'log')
                if '^' in text_calc:
                    text_calc=text_calc.replace('^', '**')
                result = eval(text_calc)

            display.delete(1.0, END)
            display.insert(1.0, str(result))
            
        except:
            display.delete(1.0, END)
            display.insert(1.0, "Error")

    elif text=='back':
        text_calc = display.get("1.0", END) 

        if text_calc.endswith('sqrt\n'):
            display.delete("end-5c", END)

        elif text_calc.endswith('pi\n'):
            display.delete("end-3c", END)

        elif text_calc.endswith('sin\n' or 'cos\n' or 'log\n'):
            display.delete("end-4c", END)

        else:
            display.delete("end-2c", END)
    else:
        display.insert(END, text)

from tkinter import *
from numpy import *

#######################################################################################
root = Tk()
root.title("파이썬 계산기")

#########################################################################################

display = Text(root, width=30, height=3, borderwidth=5)
display.grid(row=0, column=0, columnspan=6, padx=10, pady=10)

############################################################################################
buttons = [
    ('(', 1, 0), (')', 1, 1), ('C', 1, 2), ('back', 1, 3), ('sin', 1, 4), ('^', 1, 5),
    ('7', 2, 0), ('8', 2, 1), ('9', 2, 2), ('/', 2, 3), ('cos', 2, 4), ('sqrt', 2, 5),
    ('4', 3, 0), ('5', 3, 1), ('6', 3, 2), ('*', 3, 3), ('tan', 3, 4), ('!', 3, 5),
    ('1', 4, 0), ('2', 4, 1), ('3', 4, 2), ('-', 4, 3), ('log10', 4, 4), ('e', 4, 5),
    ('0', 5, 0), ('.', 5, 1), ('=', 5, 2), ('+', 5, 3), ('ln', 5, 4), ('pi', 5, 5)
]

for text, row, col in buttons:
    button = Button(root, text=text, width=5, height=4, 
                    command = lambda t=text: button_click(t))
    button.grid(row=row, column=col, padx=5, pady=5)

root.mainloop()
```
