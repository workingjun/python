```python
import requests
from bs4 import BeautifulSoup


page_num = 1
keyword = "갤럭시 z폴드 3케이스" #input()


while page_num<2:
        
        coupang_url = 'https://www.coupang.com/np/search?q='+ keyword + '&channel=user&component=&eventCategory=SRP&trcid=&traid=&sorter=scoreDesc&minPrice=&maxPrice=&priceRange=&filterType=&listSize=36&filter=&isPriceRange=false&brand=&offerCondition=&rating=0&page='+ str(page_num)+'&rocketAll=false&searchIndexingToken=1=9&backgroundColor='

        header={'User-Agent':"Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/121.0.0.0 Safari/537.36", 
        "Accept-Language": "ko-KR,ko;q=0.9"}

        raw = requests.get(coupang_url, headers=header) 
        search = BeautifulSoup(raw.text, 'html.parser')


        box = search.find('ul', {'id' : 'productList'})
        all_product = box.find_all('li', {'class': 'search-product'})

        print('페이지:', page_num, "<쿠팡" + keyword + "결과>:")

        for i, product in enumerate(all_product):
                if i==0:
                     continue
                try:
                        imagefind = product.find("dt", {'class' : 'image'})
                        imageLink = "http:" + imagefind.find("img", {'class' : 'search-product-wrap-img'})['src']

                        if "img1a" in imageLink:
                               continue 

                        title = product.find('div', {'class' : 'name'})
                        price = product.find('strong', {'class' : 'price-value'})
                        point = product.find('div', {'class' : 'other-info'}).find('span', {'class' : 'rating-total-count'})
                        link2 = product.find('a')['href']
                        link = "https://www.coupang.com" + link2
                        
                        print(i, '.')
                        print("이미지 링크:", imageLink)
                        print('이름 : ', title.text.strip())
                        print('가격 : ', price.text)
                        print('별점 : ', point.text)
                        print('링크:', link)
                        print()
                except:
                       pass        

        print()
        page_num+=1
```

```python
from tkinter import *
from PIL import Image, ImageTk
from urllib.request import urlopen
from io import BytesIO
from coupang_search import Search 

num=0

def delete_and_insert():
    global imageLink_list
    global title_list
    global price_list
    global point_list

    display1.delete(1.0, END)
    display2.delete(1.0, END)
    display3.delete(1.0, END)
    display4.delete(1.0, END)

    display1.insert(1.0, title_list[num])
    display2.insert(1.0, price_list[num])
    display3.insert(1.0, point_list[num])
    display4.insert(1.0, link_list[num])

def img_insert():
    global link_list

    image_url = imageLink_list[num]

    with urlopen(image_url) as response:
        image_data = response.read()
        image = Image.open(BytesIO(image_data))
        photo = ImageTk.PhotoImage(image)

    lable1.configure(image=photo)
    lable1.image=photo
    
def clickNext():
    global num
    global imageLink_list

    num+=1
    if num>len(imageLink_list):
        num=0

    img_insert()
    
    delete_and_insert()
    
def clickPrev():
    global num
    global imageLink_list

    num-=1
    if num<0:
        num = len(imageLink_list)
    
    img_insert()
    
    delete_and_insert()

def search_action():
    global link_list
    global imageLink_list
    global title_list
    global price_list
    global point_list
    #global keyword

    keyword = keyword_entry.get()

    search_instance = Search(keyword)
    search_instance.elem_search()

    imageLink_list = search_instance.imageLink_list
    title_list = search_instance.title_list
    price_list = search_instance.price_list
    point_list = search_instance.point_list
    link_list = search_instance.link_list

win=Tk()
win.title("coupang shopping")
win.geometry("713x600")

information = Label(win, text="아래 검색어를 입력한 뒤, 검색 버튼을 누르세요.")
information.grid(column=1, row=0)

keyword_entry = Entry(win, width=60)
keyword_entry.grid(column=1, row=1)

search_button = Button(win, text="검색", command=lambda: [search_action(), clickNext()])
search_button.grid(column=2, row=1)

btnPrev=Button(win, text="<<이전", command=clickPrev)
btnNext=Button(win, text="다음>>", command=clickNext)

keyword="검색어"

information_list = Label(win, text=f"<{keyword}의 쇼핑 목록>")
information_list.grid(column=1, row=2)

lable1 = Label(win)
display1 = Text(win, width=100, height=3, borderwidth=5)
display2 = Text(win, width=100, height=3, borderwidth=5)
display3 = Text(win, width=100, height=3, borderwidth=5)
display4 = Text(win, width=100, height=3, borderwidth=5)


lable1.grid(column=0, row=3, columnspan=3)

display1.grid(column=0, row=4, columnspan=3)
display2.grid(column=0, row=5, columnspan=3)
display3.grid(column=0, row=6, columnspan=3)
display4.grid(column=0, row=7, columnspan=3)

btnPrev.grid(column=0, row=8)
btnNext.grid(column=2, row=8)

win.mainloop()
```
