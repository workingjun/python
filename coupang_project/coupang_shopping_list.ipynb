```python
import requests
from bs4 import BeautifulSoup
######################################## 쿠팡 검색하기 ######################################################
## 페이지 수 
page_num = 1

## 키워드 입력 
keyword = "노트북" ##input() 

## 페이지 수만큼 반복 
while page_num<2:

        ## 홈페이지 주소
        base_url = "https://www.coupang.com/np/search"
        ##추가할 옵션
        query_params = [
        f"q={keyword}",
        "channel=user",
        "component=",
        "eventCategory=SRP",
        "trcid=",
        "traid=",
        "sorter=scoreDesc",
        "minPrice=",
        "maxPrice=",
        "priceRange=",
        "filterType=",
        "listSize=36",
        "filter=",
        "isPriceRange=false",
        "brand=",
        "offerCondition=",
        "rating=0",
        f"page={page_num}",
        "rocketAll=false",
        "searchIndexingToken=1=9",
        "backgroundColor="
        ]

        ## 문자열 합치기
        coupang_url = f"{base_url}?" + "&".join(query_params)

        ## 서버에 요청 보낼 때 거부 당하기 않기 위해 헤더 붙이기 
        ## 안에 문자들은 임의의 값이 아님
        ## 그 사이트가 정해준 암호 
        header={
                'User-Agent': (
                                "Mozilla/5.0 (Windows NT 10.0; Win64; x64) " 
                                "AppleWebKit/537.36 (KHTML, like Gecko) " 
                                "Chrome/121.0.0.0 Safari/537.36"
                                ), 
                "Accept-Language": "ko-KR,ko;q=0.9"
                }

        ## 요청 보내기
        raw = requests.get(coupang_url, headers=header) 

        ## html을 입력으로 넣고 파싱함 -> 구조 구문 분석 
        search = BeautifulSoup(raw.text, 'html.parser')
        ########################################################## 여기까진 공식 

        ## 태그 네임이 ul이고 id가 productList인 요소 찾기
        box = search.find('ul', {'id' : 'productList'})

        ## 태그 네임이 li인 모든 요소들 찾기 -> 결과는 변수에 리스트로 저장됨
        all_product = box.find_all('li', {'class': 'search-product'})

        print('페이지:', page_num, "<쿠팡" + keyword + "결과>:")

        
        for i, product in enumerate(all_product):

                ## 이미지 링크 찾기 
                imagefind = product.find("img", {'class' : 'search-product-wrap-img'})

                ## 예외처리
                if imagefind.get("data-img-src") is None:
                        imageLink = "http:" + imagefind.get('src')
                else:
                       imageLink = "http:" + imagefind.get("data-img-src")

                ## 제목
                title = product.find('div', {'class' : 'name'})

                ## 가격
                price = product.find('strong', {'class' : 'price-value'})

                ## 별점
                point = product.find('span', {'class' : 'rating-total-count'})
                
                ## 별점 예외처리
                if point is None:
                       point=''

                ## 링크 가공
                link2 = product.find('a')['href']
                link = "https://www.coupang.com" + link2
                
                print(i, '.')
                print("이미지 링크:", imageLink)
                print('이름 : ', title.text.strip())
                print('가격 : ', price.text)
                
                if isinstance(point, str):
                        print('별점 : ', point)
                else:
                       print('별점 : ', point.text)

                print('링크:', link)
                print()
                      
        print()
        page_num+=1
```
