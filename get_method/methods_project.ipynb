```python
##메소드 가져오기 
def get_methods(obj):
    methods = [method for method in dir(obj) if not method.endswith('__')]
    return methods

st = get_methods(str)
inta=get_methods(int)
flo=get_methods(float)

di = get_methods(dict)
se = get_methods(set)
tu = get_methods(tuple)

li = get_methods(list)
```

```python
import io
import sys


def method_description(name, obj):
    

    help_strings = []

    for method_name in name:
        method = getattr(obj, method_name)
        
        # 문자열 버퍼 생성
        buffer = io.StringIO()
        
        # 이전의 sys.stdout 저장
        old_stdout = sys.stdout
        
        # help() 함수의 출력을 문자열 버퍼에 캡처
        sys.stdout = buffer
        help(method)
        
        # 이전의 sys.stdout으로 복원
        sys.stdout = old_stdout
        
        # 문자열 버퍼의 내용을 리스트에 추가
        help_strings.append(buffer.getvalue())
        buffer.close()
    return help_strings
```

```python
st1= method_description(st, str)
inta1= method_description(inta, int)
flo1= method_description(flo, float)

se1= method_description(se, set)
di1= method_description(di, dict)
tu1= method_description(tu, tuple)

li1= method_description(li, list)
```

```python
from bs4 import BeautifulSoup   
import requests
```

```python
import pandas as pd
import numpy as np
```

```python
url='https://dojang.io/mod/page/view.php?id=2464'

raw = requests.get(url)
soup = BeautifulSoup(raw.text, 'html.parser')
```

```python
table=soup.find('div', 'no-overflow').find_all('tr')
table = table[1:]

```

```python
func=[]
descri=[]

for elem in table:
    func.append(elem.find('span', 'cdio-codei').get_text())
    
    des_table = elem.find_all('p')
    des_table=des_table[1:]
    
    text=''
    for elem2 in des_table:
        text+=elem2.get_text()
        
    descri.append(text)
```

```python
# 구글번역 라이브러리를 이용해 번역
import googletrans

def translate(text):
    translator = googletrans.Translator()

    outStr=[]

    for i in text:
        a=translator.translate(i, dest = 'ko', src = 'en')
        outStr.append(a.text)
    return outStr
```

```python
inta1=translate(inta1)
flo1=translate(flo1)
st1=translate(st1)

tu1=translate(tu1)
se1=translate(se1)
di1=translate(di1)

li1=translate(li1)
```

```python
max_val = max(len(st1), len(inta1), len(flo1), len(se1), len(di1), len(tu1), len(li1))
```

```python
df=pd.DataFrame()



df['내장함수']=func + [np.NaN]*(max_val-len(func))
df['내장함수 설명']=descri+ [np.NaN]*(max_val-len(func))

df['정수형']=inta + [np.NaN]*(max_val-len(inta))
df['정수형 설명']=inta1+ [np.NaN]*(max_val-len(inta))


df['실수형']=flo+ [np.NaN]*(max_val-len(flo))
df['실수형 설명']=flo1+ [np.NaN]*(max_val-len(flo))


df['문자열']=st+ [np.NaN]*(max_val-len(st))
df['문자열 설명']=st1+ [np.NaN]*(max_val-len(st))


df['리스트']=li+ [np.NaN]*(max_val-len(li))
df['리스트 설명']=li1+ [np.NaN]*(max_val-len(li))


df['튜플']=tu+ [np.NaN]*(max_val-len(tu))
df['튜플 설명']=tu1+ [np.NaN]*(max_val-len(tu))


df['세트']=se+ [np.NaN]*(max_val-len(se))
df['세트 설명']=se1+ [np.NaN]*(max_val-len(se))


df['딕셔너리']=di+ [np.NaN]*(max_val-len(di))
df['딕셔너리 설명']=di1+ [np.NaN]*(max_val-len(di))

```

```python
df.to_csv("파일이름.csv",encoding="utf-8-sig", index=False)
```
