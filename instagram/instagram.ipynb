```python
from dotenv import load_dotenv
import os

# .env 파일 로드
load_dotenv()

my_id = os.getenv('my_id')
my_password = os.getenv('my_password')
```

```python
from selenium import webdriver
import chromedriver_autoinstaller
from selenium.webdriver.common.by import By
from time import sleep
from selenium.common.exceptions import NoSuchElementException

chromedriver_autoinstaller.install()
driver=webdriver.Chrome()
driver.implicitly_wait(6)

login_url= 'https://www.instagram.com/explore/tags/비키니/'
driver.get(url=login_url)

sleep(2)
get_ele = driver.find_element(By.XPATH, '/html/body/div[2]/div/div/div[2]/div/div/div[1]/section/nav/div[2]/div/div/div[2]/div/div/div[2]/div[1]/a')

get_link=get_ele.get_attribute("href")
driver.get(url=get_link)


input_Id=driver.find_element(By.XPATH, '//*[@id="loginForm"]/div/div[1]/div/label/input')
input_Id.send_keys(my_id)

input_password = driver.find_element(By.XPATH, '//*[@id="loginForm"]/div/div[2]/div/label/input')
input_password.send_keys(my_password)

driver.find_element(By.XPATH, '//*[@id="loginForm"]/div/div[3]/button').click()
driver.find_element(By.XPATH, '/html/body/div[2]/div/div/div[2]/div/div/div[1]/div[1]/div[2]/section/main/div/div/div/div/div').click()

#for _ in range(5):
    #driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    #sleep(5)  # 시간 대기 후 페이지 로딩이 완료되었는지 확인

link_list=[]
title=[]

xpath=f'/html/body/div[2]/div/div/div[2]/div/div/div[1]/div[1]/div[2]/section/main/article/div/div[2]/div/div[{1}]/div[{1}]'
driver.find_element(By.XPATH, xpath).click()

def link_title_save():
    global link_list, title
    xpath = "//div[@class='_aagu _aato']"
    elem = driver.find_element(By.XPATH, xpath)

    link = elem.find_element(By.TAG_NAME, 'img')
    link1 = link.get_attribute('src')
    link_list.append(link1)

    xpath="//div[@class='_aaqt']"
    elem = driver.find_element(By.XPATH, xpath)
    title_name = elem.find_element(By.TAG_NAME, 'a').text
    title.append(title_name)

num=5

for i in range(num):
    
    link_title_save()
    print(1)
    while 1:
        try:
            xpath = "//button[@class=' _afxw _al46 _al47']"
            elem = driver.find_element(By.XPATH, xpath)

            elem.click()

            link_title_save()
        except NoSuchElementException:
            break
    
    xpath2="//div[@class=' _aaqg _aaqh']"
    elem2 = driver.find_element(By.XPATH, xpath2).click()

    sleep(2)
```

```python
import requests
import os
from urllib.parse import urlparse, unquote


def download_image(url, save_path):
    # 이미지 데이터를 요청
    response = requests.get(url)

    # 이미지 데이터를 파일로 저장
    with open(save_path, 'wb') as file:
        file.write(response.content)
    print(f"이미지가 성공적으로 저장되었습니다: {save_path}")

img_num=1

for i in range(len(link_list)):
    image_url = link_list[i]  # 다운로드 받을 이미지 URL
    save_directory = "비키니"  # 이미지를 저장할 디렉토리
    os.makedirs(save_directory, exist_ok=True)  # 디렉토리가 존재하지 않으면 생성

    
    if title[i]==title[i-1]:
        image_name = f"{title[i]}_{img_num}.jpg"    
        img_num += 1
    else:
        image_name = f"{title[i]}.jpg"  # URL 인코딩된 문자열을 디코딩 (선택 사항)

    # 디렉토리와 파일 이름을 합쳐 저장할 전체 경로 설정
    save_path = os.path.join(save_directory, image_name)  # 여기서 save_path는 "images/image.jpg"가 됨

    download_image(image_url, save_path)
```
