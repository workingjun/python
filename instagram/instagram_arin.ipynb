```python
from selenium import webdriver
import chromedriver_autoinstaller
from selenium.webdriver.common.by import By
from time import sleep
from selenium.common.exceptions import NoSuchElementException
#from dotenv import load_dotenv
import os

# .env 파일 로드
#load_dotenv()

my_id = os.getenv('my_id')
my_password = os.getenv('my_password')

chromedriver_autoinstaller.install()
driver=webdriver.Chrome()
driver.implicitly_wait(10)

################################################################################################

login_url= 'https://www.instagram.com/explore/tags/아린/'
driver.get(url=login_url)

sleep(2)
get_ele = driver.find_element(By.XPATH, '/html/body/div[2]/div/div/div[2]/div/div/div[1]/section/nav/div[2]/div/div/div[2]/div/div/div[2]/div[1]/a')

get_link=get_ele.get_attribute("href")
driver.get(url=get_link)


input_Id=driver.find_element(By.XPATH, '//*[@id="loginForm"]/div/div[1]/div/label/input')
input_Id.send_keys(my_id)

input_password = driver.find_element(By.XPATH, '//*[@id="loginForm"]/div/div[2]/div/label/input')
input_password.send_keys(my_password)

driver.find_element(By.XPATH, '//*[@id="loginForm"]/div/div[3]/button').click()
driver.find_element(By.XPATH, '/html/body/div[2]/div/div/div[2]/div/div/div[1]/div[1]/div[2]/section/main/div/div/div/div/div').click()

##############################################################################################################################
link_list=[]
title=[]

video_list=[]

xpath=f'/html/body/div[2]/div/div/div[2]/div/div/div[1]/div[1]/div[2]/section/main/article/div/div[2]/div/div[{1}]/div[{1}]'
driver.find_element(By.XPATH, xpath).click()

def link_title_save():
    xpath = "//div[@class='_aagu _aato']"
    elem = driver.find_element(By.XPATH, xpath)

    link = elem.find_element(By.TAG_NAME, 'img')
    link1 = link.get_attribute('src')
    link_list.append(link1)

    xpath="//div[@class='_aaqt']"
    elem = driver.find_element(By.XPATH, xpath)
    title_name = elem.find_element(By.TAG_NAME, 'a').text
    title.append(title_name)

num=100

for i in range(num):
    try:
        link_title_save()

        while 1:
            try:
                xpath = "//button[@class=' _afxw _al46 _al47']"
                elem = driver.find_element(By.XPATH, xpath)

                elem.click()

                link_title_save()

            except NoSuchElementException:
                break

    except NoSuchElementException:
        xpath = "//div[@class='_aasm']"
        driver.find_element(By.XPATH, xpath).click()

        xpath = "//button[text()='링크 복사']"
        driver.find_element(By.XPATH, xpath).click()
        current_url = driver.current_url
        video_list.append(current_url)
    
    xpath2="//div[@class=' _aaqg _aaqh']"
    driver.find_element(By.XPATH, xpath2).click()

    sleep(2)
```

```python
import requests
import os
from urllib.parse import urlparse, unquote
from datetime import datetime

def download_image(url, save_path):
    # 이미지 데이터를 요청
    response = requests.get(url)

    # 이미지 데이터를 파일로 저장
    with open(save_path, 'wb') as file:
        file.write(response.content)
    print(f"이미지가 성공적으로 저장되었습니다: {save_path}")

img_num=1

for i in range(len(link_list)):
    image_url = link_list[i]  # 다운로드 받을 이미지 URL
    save_directory = "Arin2"  # 이미지를 저장할 디렉토리
    os.makedirs(save_directory, exist_ok=True)  # 디렉토리가 존재하지 않으면 생성

    current_date = datetime.now().strftime("%Y-%m-%d")

    image_name = f"{current_date}_{title[i]}_{img_num}.jpg"    
     # URL 인코딩된 문자열을 디코딩 (선택 사항)
    img_num += 1
    
    # 디렉토리와 파일 이름을 합쳐 저장할 전체 경로 설정
    save_path = os.path.join(save_directory, image_name)  # 여기서 save_path는 "images/image.jpg"가 됨

    download_image(image_url, save_path)
```

```python
import instaloader

# 인스타로더 인스턴스 생성
L = instaloader.Instaloader()

current_directory = os.getcwd()
save_directory = os.path.join(current_directory, "Arin_video")  # 이미지를 저장할 디렉토리
os.makedirs(save_directory, exist_ok=True) 

for i in range(len(video_list)):
    # 인스타그램 게시물 URL
    post_url = video_list[i].split('/')[-2]
    # 게시물 불러오기
    post = instaloader.Post.from_shortcode(L.context, post_url)
    
    L.download_post(post, target="Arin_video")

######################################################################################################
import os

def delete_files_with_extension(directory, extension):
    # 디렉토리 내의 모든 파일에 대해 반복
    for filename in os.listdir(directory):
        filepath = os.path.join(directory, filename)
        try:
            # 파일인지 확인하고 삭제
            if os.path.isfile(filepath) and filename.endswith(extension):
                os.remove(filepath)
                print(f"Deleted: {filepath}")
            # 디렉토리인 경우 재귀적으로 함수 호출
            elif os.path.isdir(filepath):
                delete_files_with_extension(filepath, extension)
        except Exception as e:
            print(f"Error deleting {filepath}: {e}")

# 삭제할 디렉토리 경로와 삭제할 확장자 지정
directory_path = 'Arin_video'
file_extension = '.json.xz'

# 함수 호출
delete_files_with_extension(directory_path, file_extension)

# 삭제할 디렉토리 경로와 삭제할 확장자 지정
directory_path = 'Arin_video'
file_extension = '.jpg'

# 함수 호출
delete_files_with_extension(directory_path, file_extension)

# 삭제할 디렉토리 경로와 삭제할 확장자 지정
directory_path = 'Arin_video'
file_extension = '.txt'

# 함수 호출
delete_files_with_extension(directory_path, file_extension)
```
