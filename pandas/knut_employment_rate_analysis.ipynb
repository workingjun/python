```python

len0 = len(df['61.5'])

f=open("D:\\취업률 정리.txt", 'w', encoding='utf-8')

f.write("="*50)
f.write("\n")
f.write('공과대학')
f.write("\n")
f.write("\n")
f.write('0 '+"건축공학전공 "+ "주간 "+ "일반과정 "+ '\n') 
f.write("졸업자: 46 "+ "\n"+ "취업률: 61.5")
f.write("\n")
f.write("\n")

sum=[]
count=[]

sum1=61.5
count1=1

for i in range(len0):
    
    if str(df['공과대학'][i])!="nan":
        sum.append(sum1)
        count.append(count1)
        sum1=0
        count1=0

        f.write("="*50)
        f.write("\n")
        f.write(str(df['공과대학'][i])+' ')
        f.write("\n")
        f.write("\n")
    
    f.write(str(i+1)+ ' ')
    f.write(str(df['건축공학전공'][i])+' ')
    f.write(str(df['주간'][i])+' ')
    f.write(str(df['일반과정'][i])+' ')
    f.write("\n")
    f.write("졸업자: "+ str(df['33'][i]+df['13'][i])+'\n')
    f.write("취업률: "+str(df['61.5'][i])) ##취업률 정보
    f.write("\n")
    f.write("\n")

    sum1+=df['61.5'][i]
    count1+=1

sum.append(sum1)
count.append(count1)

f.write("="*50)
f.write("\n")
f.write('대대별 취업률 통계')
f.write("\n")

mylist=['공과', '보건', '인문사회', '철도', '융기']

for i in range(len(sum)):
    f.write(mylist[i])
    f.write(str(sum[i]/count[i]))
    f.write("\n")

f.write("="*50)
f.write("\n")

f.close()


```
