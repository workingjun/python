```python
from matplotlib import font_manager

# 폰트 경로 지정
font_path = 'c:\\WINDOWS\\Fonts\\MALGUN.TTF'  # 사용하고자 하는 폰트의 경로를 지정합니다.
font_prop = font_manager.FontProperties(fname=font_path)

# 폰트 설정
plt.rcParams['font.family'] = font_prop.get_name()


# 그래프 생성
plt.bar(df['자치구별(2)'][2:], )
plt.xticks(rotation=90)  # x축 레이블을 45도로 회전

# 그래프 출력
plt.show()
```

```python
df2 = df.iloc[2:, 1:]

df2 = df2.rename(columns={'살인.1': '살인검거', '살인': '살인발생',
                          '강도.1': '강도검거', '강도': '강도발생',
                          '강간·강제추행.1': '강간·강제추행검거', '강간·강제추행': '강간·강제추행발생',
                          '절도.1': '절도검거', '절도': '절도발생', 
                          '폭력.1': '폭력검거', '폭력': '폭력발생'})

df2
```

```python
import numpy as np

table=pd.pivot_table(df2, index=['자치구별(2)'], 
	        values=['절도발생', '절도검거'])

table = table.sort_values(by=['절도발생', '절도검거'], ascending=False)

table.plot(kind='bar')
plt.show()
```
