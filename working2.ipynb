```python
import numpy as np

def xor_model(x1, x2):
    x = np.array([x1, x2])
    ws1 = np.array([-0.5, -0.5])
    ws2 = np.array([0.5, 0.5])
    ws3 = np.array([0.5, 0.5])

    b1 = 0.7
    b2 = -0.2
    b3 = -1.25

    s1 = np.sum(ws1 * x) + b1
    s2 = np.sum(ws2 * x) + b2
    s3 = np.sum(ws3 * x) + b3

    s1 = 0 if s1 <= 0 else 1
    s2 = 0 if s2 <= 0 else 1
    s3 = 0 if s2 <= 0 else 1


    wy = np.array([0.5, 0.5, 0.5])
    b4 = -0.7

    tmp = np.sum(wy * [s1, s2, s3]) + b4
    if tmp <= 0:
        return 0
    else:
        return 1

if __name__ == '__main__':
    # data_2D = np.array([[0, 0], [1, 0], [0, 1], [1, 1]], dtype=np.int16)
    data_2D = np.array([[0, 0], [1, 0], [0, 1], [1, 1], [1.5, 1.5]])

    for xs in data_2D:
        y = xor_model(xs[0], xs[1])
        print(str(xs) + " -> " + str(y))
        
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.fft import fft

# 입력 신호 생성
fs = 1000  # 샘플링 주파수
duration = 1  # 신호의 지속시간
num_samples = fs * duration  # 전체 샘플 수
t = np.linspace(0, duration, num_samples)##, endpoint=False)  # 시간 배열 생성

f1 = 5  # 주파수 5Hz
f2 = 50  # 주파수 50Hz
signal = np.sin(2*np.pi*f1*t)   # 복합 신호 생성

# 푸리에 변환 계산
fft_result = np.fft.fft(signal)
amplitude = abs(fft_result)*(2/len(fft_result))
# 주파수 계산
freqs = np.fft.fftfreq(len(t), 1/fs)

# 복소수 표현에서 위상 추출
phase_spectrum = np.angle(fft_result)
# 위상 스펙트럼을 0부터 2π까지의 범위로 변환
phase_spectrum = np.where(phase_spectrum < 0, 2*np.pi + phase_spectrum, phase_spectrum)

# 시각화
plt.figure(figsize=(10, 6))

plt.subplot(3, 1, 1)
plt.plot(t, signal)
plt.title('Input Signal')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')

plt.subplot(3, 1, 2)
plt.stem(freqs, amplitude)
plt.title('abs Spectrum')
plt.xlabel('Frequency (Hz)')
plt.ylabel('amplitude (Voltage)')
plt.xlim(-7,7)  # 양의 주파수 영역만 표시
plt.grid()


plt.subplot(3, 1, 3)
plt.plot(freqs, phase_spectrum)
plt.title('Phase Spectrum')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Phase (radians)')
plt.xlim(-100, fs/2)  # 양의 주파수 영역만 표시
plt.grid()

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

class NeuralNetwork:
    def __init__(self, input_size, learning_rate=0.01, epochs=10):
        self.weights = np.random.randn(input_size + 1)
        self.learning_rate = learning_rate
        self.epochs = epochs

    def sigmoid(self, x):
        return 1 / (1 + np.exp(-x))

    def sigmoid_derivative(self, x):
        return x * (1 - x)

    def train(self, X, y):
        X = np.insert(X, 0, 1, axis=1)  # 편향(bias) 추가
        for _ in range(self.epochs):
            output = self.sigmoid(np.dot(X, self.weights))
            error = y - output
            adjustment = np.dot(X.T, error * self.sigmoid_derivative(output))
            self.weights += self.learning_rate * adjustment

    def predict(self, X):
        X = np.insert(X, 0, 1, axis=1)  # 편향(bias) 추가
        return self.sigmoid(np.dot(X, self.weights))

# 입력 데이터와 출력 데이터 설정
X = np.array([[170, 80], [175, 76], [180, 70], [160, 55], [163, 43], [165, 48]])  # 신장(cm), 체중(kg)
y = np.array([0, 0, 0, 1, 1, 1])  # 1: 여성, 0: 남성

# 신경망 모델 생성 및 훈련
neural_network = NeuralNetwork(input_size=2)
neural_network.train(X, y)

# 플롯 설정
plt.figure(figsize=(8, 6))
plt.title('Classification of Gender by Height and Weight', fontsize=15)
plt.xlabel('Height (cm)', fontsize=12)
plt.ylabel('Weight (kg)', fontsize=12)

# 데이터 포인트 플롯
for i in range(len(y)):
    if y[i] == 0:
        plt.scatter(X[i][0], X[i][1], color='blue')  # 남성은 파란색으로 표시
    else:
        plt.scatter(X[i][0], X[i][1], color='red')  # 여성은 빨간색으로 표시

# 결정 경계 플롯
w0, w1, w2 = neural_network.weights
x_values = np.array([150, 190])
y_values = (-w0 +w1 * x_values) / w2
plt.plot(x_values, y_values, color='green', linestyle='-', label='Decision Boundary')  # 결정 경계 플롯

plt.xticks([])  # x축 눈금 없애기
plt.yticks([])  # y축 눈금 없애기
plt.grid(False)  # 그리드 없애기
plt.legend()  # 범례 추가
plt.show()
```
