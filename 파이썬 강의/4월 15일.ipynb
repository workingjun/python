```python
class students:
    def __init__(self):
        print("학번 입력:")
        self.number=int(input())
        print("이름 입력:")
        self.name=input()
        print("평점 입력:")
        self.point=int(input())
    
    def save(self):
        global list1
        list1.append(f'학번: {self.number}, 이름: {self.name}, 점수: {self.point}')
    
    def __str__(self):
        global list1
        sum_str=''

        for i in list1:
            sum_str += i + '\n'
            
        return f"{sum_str}"
        
```

```python
class Calculator:
    def __init__(self, a, b):
        self.a=a
        self.b=b
        self.value = 0

    def add(self):
        self.value = a + b
        return self.value
    
    def show(self):
        print(f"입력된 두 수: {a}, {b}")

class MaxlLimitCalculator(Calculator):
    def limit(self):
        if self.value>=100:
            self.value=100
        return self.value

    def __str__(self):
        return f"{self.value}"
    

a=int(input())
b=int(input())

add_calc=MaxlLimitCalculator(a, b)
add_calc.show()
add_calc.add()
add_calc.limit()

print(f'두 수의 합: {add_calc}')

# add_calc=Calculator(a, b)
# add_calc.show()
# add_calc.add()
# add_calc.limit()

# print(f'두 수의 합: {add_calc}')
# ## AttributeError: 'Calculator' object has no attribute 'limit'
```
